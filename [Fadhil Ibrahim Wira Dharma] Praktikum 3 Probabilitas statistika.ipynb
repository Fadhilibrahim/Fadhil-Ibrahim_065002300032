```python
import pandas as pd

data_fadhil = pd.read_clipboard()
print(data_fadhil)
```

```python
subset_tinggi_160 = data_fadhil[data_fadhil['Tinggi Badan'] == 160]
print(subset_tinggi_160)
```

```python
subset_tinggi_lebih_besar_dari_170 = data_fadhil[data_fadhil['Tinggi Badan'] > 170]
print(subset_tinggi_lebih_besar_dari_170)
```

```python
data_fadhil.loc[data_fadhil['Tinggi Badan'] > 160, 'Tinggi Badan'] = 'Tinggi'
data_fadhil.loc[data_fadhil['Tinggi Badan'] != 'Tinggi', 'Tinggi Badan'] = 'Pendek'

print(data_fadhil)
```

```python
data_fadhil['Jurusan'] = 'Infor20'
data_fadhil['Fakultas'] = 'FTI'

print(data_fadhil)
```

```python
data_fadhil = data_fadhil.drop('Fakultas', axis=1)
print(data_fadhil)
```

```python
data_fadhil['Nama_Gender'] = data_fadhil['Nama'].str.cat(data_fadhil['Gender'], sep='_')
data_fadhil['Angkatan_TinggiBadan'] = data_fadhil['Angkatan'].astype(str) + '_' + data_fadhil['Tinggi Badan'].astype(str)
data_fadhil['Gabungan'] = data_fadhil['Nama_Gender'].str.cat(data_fadhil['Angkatan_TinggiBadan'], sep='_')
print(data_fadhil)
```

```python
gabungan_baris_1_5 = data_fadhil.iloc[0:5]
gabungan_baris_25_30 = data_fadhil.iloc[24:30]
gabungan_total = pd.concat([gabungan_baris_1_5, gabungan_baris_25_30])

print(gabungan_total)
```

```python
data_fadhil_sorted = data_fadhil.sort_values(by='Waktu Perjalanan')
print(data_fadhil_sorted)
```
